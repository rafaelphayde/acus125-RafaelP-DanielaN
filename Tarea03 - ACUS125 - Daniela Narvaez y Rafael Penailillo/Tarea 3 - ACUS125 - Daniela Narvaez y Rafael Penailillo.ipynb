## Ejercicio propuesto:
> + Graficar distintos abistamientos de sismo ocurrido el 14/05/2015 a las 15:08:01, de Magnitud 5,2

```python
import obspy
from obspy.core import read
import matplotlib.pyplot as plt
import os
from os.path import join
```

```python
path = r'2015_05_14 15_08_01_M5_2'
```

```python
nombres_carpetas = os.listdir(path)
carpetas = [os.path.join(path, carpeta) for carpeta in nombres_carpetas]
```

```python
def grafica(archivos: list) -> list:
    ''' Función para graficar las señales de cada carpeta, recibe una lista
    con los path de cada archivo, las grafica  y retorna el máximo de cada eje'''

    estacion = obspy.read(archivos[0])
    estacion += obspy.read(archivos[1])
    estacion += obspy.read(archivos[2])

    estacion[0].plot();
    estacion[1].plot();
    estacion[2].plot();
    
    return int(estacion[0].max()), int(estacion[1].max()), int(estacion[2].max())
```

```python
directorios={}
```

```python
for indx, carpeta in enumerate(carpetas):
    archivos = os.listdir(carpeta)
    archivos = [os.path.join(carpeta, archivo) for archivo in archivos]
    directorios[nombres_carpetas[indx]] = archivos
```

```python
directorios['CO02']
```

```python
estacion = obspy.read('2015_05_14 15_08_01_M5_2\\CO02\\CO02_HHE.sac')
```

## Estación CO02

```python
maxE, maxN, maxZ = grafica(directorios['CO02'])
```

De estos tres gráficos, se obtienen los máximos en cada eje, los cuales son:

```python
print(f'Eje Este: {maxE} \nEje Norte: {maxN} \nEje Z: {maxZ}')
```

Esta estación presentó el máximo peak en el eje Norte a 319139 counts hacia el sur
